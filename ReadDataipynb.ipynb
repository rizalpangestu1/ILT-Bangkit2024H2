```python
import zipfile

with zipfile.ZipFile('animal-image-dataset-cats-dogs-and-foxes.zip', 'r') as zip_ref:
    zip_ref.extractall('animal-image-dataset-cats-dogs-and-foxes')
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Define the main directory of the dataset
data_dir = '/content/animal-image-dataset-cats-dogs-and-foxes/Animal Image Dataset-Cats, Dogs, and Foxes'

# Function to count images in each class and show sample images
def explore_dataset(data_dir):
    classes = os.listdir(data_dir)
    class_counts = {}

    # Count images and display one sample per class
    for cls in classes:
        cls_path = os.path.join(data_dir, cls)
        if os.path.isdir(cls_path):
            image_files = os.listdir(cls_path)
            class_counts[cls] = len(image_files)
            print(f"Class '{cls}' has {len(image_files)} images.")

            # Display a sample image for the class
            sample_image_path = os.path.join(cls_path, image_files[0])
            img = Image.open(sample_image_path)
            plt.imshow(img)
            plt.title(f"Sample from class: {cls}")
            plt.axis('off')
            plt.show()

    total_images = sum(class_counts.values())
    print(f"Total classes: {len(classes)}")
    print(f"Total images in dataset: {total_images}")
    return class_counts

# Explore the dataset
class_counts = explore_dataset(data_dir)
```

```python
import zipfile
import os

# Define the path to the zip file on Google Drive
zip_path = '/content/drive/MyDrive/zoo_animals.zip'

# Extract the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('/content/zoo_animals')

# Verify extraction
extracted_files = os.listdir('/content/zoo_animals')
print("Extracted files:", extracted_files)
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Define the main dataset directory
dataset_dir = '/content/zoo_animals/zoo animals'

# Function to list classes and count images
def list_classes_and_counts(data_dir):
    classes = os.listdir(data_dir)
    class_counts = {}

    for cls in classes:
        cls_path = os.path.join(data_dir, cls)
        if os.path.isdir(cls_path):
            image_count = len(os.listdir(cls_path))
            class_counts[cls] = image_count
            print(f"Class '{cls}' has {image_count} images.")

    total_images = sum(class_counts.values())
    print(f"\nTotal classes: {len(classes)}")
    print(f"Total images in the dataset: {total_images}")
    return class_counts

# List classes and count images
class_counts = list_classes_and_counts(dataset_dir)
```

```python
# Function to display one sample image from each class
def display_sample_images(data_dir):
    classes = os.listdir(data_dir)
    plt.figure(figsize=(15, 10))  # Adjust the figure size as needed

    # Loop through each class and display one image
    for i, cls in enumerate(classes[:19]):  # Display up to 19 classes
        cls_path = os.path.join(data_dir, cls)
        if os.path.isdir(cls_path):
            # Get the first image file in the class directory
            image_file = os.listdir(cls_path)[0]
            image_path = os.path.join(cls_path, image_file)
            img = Image.open(image_path)

            # Display the image with its class name
            plt.subplot(4, 5, i + 1)  # Adjust subplot layout as needed
            plt.imshow(img)
            plt.title(cls)
            plt.axis('off')

    plt.tight_layout()
    plt.show()

# Display sample images from each class
display_sample_images(dataset_dir)
```
